```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API"),  # This is the default and can be omitted
)
```

### Sa√≠da em Formato de Tabela

```python
system_prompt = """
Voc√™ √© um historiador especializado em civiliza√ß√µes antigas.
Sempre responda no formato de tabela:

| Civiliza√ß√£o | Per√≠odo | Contribui√ß√£o Marcante | Regi√£o |
|-------------|---------|-----------------------|--------|
"""

user_prompt = "Crie uma tabela comparando 4 civiliza√ß√µes antigas (Egito, Mesopot√¢mia, Gr√©cia, Roma)."

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    
    temperature=0.5
)

print(response.choices[0].message.content)
```
