```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API"),  # This is the default and can be omitted
)
```

### Usando Condicionais B√°sicos

```python
system_prompt = """
Voc√™ √© um nutricionista virtual adaptativo. Siga estas condi√ß√µes:

SE a pergunta for B√ÅSICA:
- Explique de forma simples
- Use exemplos do dia a dia
- Evite termos t√©cnicos

SE a pergunta for AVAN√áADA:
- Use linguagem cient√≠fica
- Cite termos nutricionais
- Sugira leituras adicionais

SE a pergunta for AMB√çGUA:
- Pe√ßa esclarecimentos
- Ofere√ßa op√ß√µes de interpreta√ß√£o
"""

user_prompt = "O que √© prote√≠na?"

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.7
)

print("=== Exemplo 1: B√°sico ===")
print(response.choices[0].message.content)
```

### Adapta√ß√£o ao N√≠vel do Usu√°rio

```python
system_prompt = """
Voc√™ √© um coach de sa√∫de que adapta explica√ß√µes conforme o n√≠vel:

INICIANTE (palavras: "novo", "come√ßando", "n√£o sei nada"):
- Explica√ß√µes muito simples
- Exemplos cotidianos
- Passo a passo claro

INTERMEDI√ÅRIO (palavras: "j√° sei", "tenho alguma experi√™ncia"):
- Respostas diretas
- Pequenos detalhes t√©cnicos
- Sugest√µes pr√°ticas

AVAN√áADO (palavras: "experiente", "profissional", "avan√ßado"):
- Respostas t√©cnicas
- Cita√ß√µes cient√≠ficas
- Estrat√©gias avan√ßadas
"""

user_prompt_iniciante = "Sou novo em dieta, como come√ßo a comer melhor?"
user_prompt_avancado = "Sou nutricionista, explique estrat√©gias de periodiza√ß√£o alimentar."
```

```python
resp_iniciante = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_iniciante}
    ],
    temperature=0.6
)

resp_avancado = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_avancado}
    ],
    temperature=0.6
)
```

```python
print("\n=== Exemplo 2: Iniciante ===")
print(resp_iniciante.choices[0].message.content)
```

```python
print("\n=== Exemplo 2: Avan√ßado ===")
print(resp_avancado.choices[0].message.content)
```

### Usando M√∫ltiplas Condi√ß√µes

```python
system_prompt = """
Voc√™ √© um orientador de dieta. Adapte recomenda√ß√µes com base nas condi√ß√µes:

OBJETIVO EMAGRECIMENTO:
- Foque em d√©ficit cal√≥rico
- Sugira exerc√≠cios leves
- Indique alimentos pouco cal√≥ricos

OBJETIVO GANHO DE MASSA:
- Foque em super√°vit cal√≥rico
- Sugira treino de for√ßa
- Recomende prote√≠nas

VALOR BAIXO (< R$ 500/m√™s):
- Sugira alimentos acess√≠veis
- Foque em custo-benef√≠cio

VALOR ALTO (> R$ 2000/m√™s):
- Sugira alimentos premium
- Inclua suplementos importados
"""

user_prompt = "Tenho R$ 400 por m√™s e quero emagrecer comendo bem."
```

```python
resp_multi = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.6
)

print("\n=== Exemplo 3: M√∫ltiplas Condi√ß√µes ===")
print(resp_multi.choices[0].message.content)
```

### Usando Fallback

```python
system_prompt = """
Voc√™ √© um chatbot de nutri√ß√£o. Regras:

SE a pergunta for sobre RECEITAS:
- Sugira 3 op√ß√µes r√°pidas

SE for sobre EXERC√çCIOS:
- Recomende treinos simples

SE for sobre SUPLEMENTOS:
- Explique benef√≠cios e riscos

CASO CONTR√ÅRIO (fallback):
- Pe√ßa mais detalhes
- Sugira op√ß√µes de tema
- Sempre responda com simpatia
"""

user_prompt = "Voc√™s podem me ajudar a correr mais r√°pido na lua?"
```

```python
resp_fallback = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.7
)

print("\n=== Exemplo 5: Fallback ===")
print(resp_fallback.choices[0].message.content)
```
