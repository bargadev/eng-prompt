### Quebrar Tarefas em MÃºltiplos Passos

```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API"),  # This is the default and can be omitted
)
```

#### Multi Step BÃ¡sico

```python
system_prompt = """
VocÃª Ã© um guia de trekking ultraleve. Siga a metodologia multi-step:
1) Liste e some os pesos por categoria
2) Calcule o peso-base (sem comida/Ã¡gua/combustÃ­vel)
3) Defina a meta de reduÃ§Ã£o (15% do peso-base)
4) Proponha trocas/reduÃ§Ãµes para alcanÃ§ar a meta
5) Verifique consistÃªncia e riscos (clima/seguranÃ§a)
Formate cada passo com tÃ­tulos claros.
"""

user_prompt = """
PLANEJAMENTO (2 dias, clima ameno, sem chuva prevista)
ITENS E PESOS (g):
â€¢ Abrigo (barraca): 1450
â€¢ Saco de dormir: 900
â€¢ Isolante: 420
â€¢ Mochila: 1100
â€¢ Fogareiro: 85
â€¢ Panela titÃ¢nio: 120
â€¢ Garrafa 1L (vazia): 110
â€¢ Kit primeiros socorros: 160
â€¢ Capa de chuva: 180
â€¢ Fleece: 320
â€¢ Camiseta extra: 110
â€¢ Meias extra: 60
â€¢ Lanterna: 90
â€¢ Power bank: 210
â€¢ Kit higiene: 140

META: reduzir o peso-base em 15% mantendo seguranÃ§a e conforto tÃ©rmico.
PASSOS: execute 1â†’5 conforme metodologia.
"""
```

```python
resp = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[{"role": "system", "content": system_prompt},
              {"role": "user", "content": user_prompt}],
    temperature=0.25
)
print("=== Multi Step BÃ¡sico ===")
print(resp.choices[0].message.content)
```

### ComparaÃ§Ã£o Passo Ãšnico x Multi Passo

```python
# Single Step
single_prompt = """
Crie um plano para instalar painÃ©is solares residenciais para uma casa de 3 quartos.
Considere orÃ§amento moderado e retorno em 6-8 anos.
"""

single = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[{"role": "user", "content": single_prompt}],
    temperature=0.5
)
print("\n=== Single Step ===")
print(single.choices[0].message.content)
```
