```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API"),  # This is the default and can be omitted
)
```

### Complemento Textual
- **Expans√£o de t√≥picos** ‚Üí desenvolvimento de pontos em par√°grafos
- **Expans√£o criativa** ‚Üí narrativas envolventes
- **Expans√£o argumentativa** ‚Üí justificativas s√≥lidas
- **Expans√£o t√©cnica** ‚Üí explica√ß√µes detalhadas e exemplos pr√°ticos
- **Expans√£o de outline** ‚Üí transforma√ß√£o de estrutura em conte√∫do completo

### Expans√£o de t√≥picos

```python
system_prompt = """
Voc√™ √© um especialista em comunica√ß√£o corporativa.
Expanda t√≥picos curtos em par√°grafos claros que:
1) Desenvolvam cada ideia com exemplos reais
2) Mantenham fluidez e coer√™ncia
3) Sejam √∫teis para profissionais em cargos de lideran√ßa
"""

user_prompt = """
Expanda estes t√≥picos sobre lideran√ßa moderna:

‚Ä¢ Comunica√ß√£o transparente
‚Ä¢ Gest√£o de equipes h√≠bridas
‚Ä¢ Incentivo √† inova√ß√£o
"""

response_topics = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.5
)

print("\n=== EXPANS√ÉO DE T√ìPICOS ===")
print(response_topics.choices[0].message.content)
```

### Expans√£o Criativa

```python
system_prompt = """
Voc√™ √© um escritor criativo. Expanda ideias simples criando:
1) Cen√°rio v√≠vido
2) Personagens interessantes
3) Atmosfera envolvente
"""

user_prompt = """
Expanda esta ideia em uma micro-hist√≥ria:
"Um barista descobre um bilhete misterioso escondido dentro de um saco de caf√©."
"""

response_creative = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.8
)

print("\n=== EXPANS√ÉO CRIATIVA ===")
print(response_creative.choices[0].message.content)
```

### Expans√£o Argumentativa

```python
system_prompt = """
Voc√™ √© um consultor estrat√©gico.
Expanda argumentos b√°sicos incluindo:
1) Justificativas bem fundamentadas
2) Exemplos e dados de apoio
3) Considera√ß√£o de contrapontos
4) Conclus√£o persuasiva
"""

user_prompt = """
Expanda este argumento:
"As empresas devem adotar a semana de trabalho de 4 dias."
"""

response_argument = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.3
)

print("\n=== EXPANS√ÉO ARGUMENTATIVA ===")
print(response_argument.choices[0].message.content)
```

### Expans√£o T√©cnica

```python
system_prompt = """
Voc√™ √© um instrutor t√©cnico especializado em ci√™ncia de dados.
Expanda conceitos explicando:
1) Como funciona na pr√°tica
2) Exemplos de c√≥digo quando √∫til
3) Casos de uso aplicados em empresas
4) Boas pr√°ticas de implementa√ß√£o
"""

user_prompt = """
Expanda este conceito t√©cnico:
"Aplica√ß√£o de modelos preditivos em an√°lise de churn de clientes."
"""

response_technical = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.4
)

print("\n=== EXPANS√ÉO T√âCNICA ===")
print(response_technical.choices[0].message.content)
```

### Expans√£o de Outline

```python
outline_prompt = """
Expanda este outline em um artigo curto e completo:

T√çTULO: "Tend√™ncias em Educa√ß√£o Digital para 2026"

OUTLINE:
1. Introdu√ß√£o - Acelerando a transforma√ß√£o educacional
2. Plataformas personalizadas de aprendizagem
3. Realidade aumentada e gamifica√ß√£o
4. Aprendizado cont√≠nuo para profissionais
5. Desafios e riscos
6. Conclus√£o - O futuro da educa√ß√£o
"""

response_outline = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[{"role": "user", "content": outline_prompt}],
    temperature=0.6
)

print("\n=== EXPANS√ÉO DE OUTLINE ===")
print(response_outline.choices[0].message.content)
```
