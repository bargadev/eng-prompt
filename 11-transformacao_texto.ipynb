```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API"),  # This is the default and can be omitted
)
```

- **Transforma√ß√£o de formato** ‚Üí mudan√ßa de estrutura
- **Transforma√ß√£o de tom** ‚Üí ajuste de estilo
- **Transforma√ß√£o de perspectiva** ‚Üí mudan√ßa de ponto de vista
- **Adapta√ß√£o para p√∫blico** ‚Üí linguagem adequada ao leitor

### Transforma√ß√£o de Formato

```python
system_prompt = """
Voc√™ √© um especialista em formata√ß√£o de conte√∫do.
Transforme textos entre diferentes estruturas garantindo:
1) Clareza
2) Preserva√ß√£o das informa√ß√µes originais
3) Boa organiza√ß√£o
"""

user_prompt = """
Transforme este par√°grafo em uma receita passo a passo:

"Para preparar um bom caf√© gelado, voc√™ deve moer gr√£os de caf√© fresco, preparar um caf√© forte,
deixar esfriar, adicionar gelo em um copo alto e completar com leite ou creme a gosto.
Finalize com um toque de a√ß√∫car ou xarope de baunilha."
"""

response_format = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.3
)

print("=== RESULTADO (Formato) ===")
print(response_format.choices[0].message.content)
```

### Transforma√ß√£o de Tom

```python
system_prompt = """
Voc√™ √© um especialista em adapta√ß√£o de estilo.
Transforme textos mudando o tom sem perder o sentido.
"""

user_prompt = """
Transforme este texto jornal√≠stico em um post leve para redes sociais:

"A cidade de Lisboa recebeu mais de 4 milh√µes de turistas em 2024,
consolidando-se como um dos destinos mais procurados da Europa."
"""

response_tone = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.6
)

print("\n=== RESULTADO (Tom) ===")
print(response_tone.choices[0].message.content)
```

### Transforma√ß√£o de Ponto de Vista

```python
system_prompt = """
Voc√™ √© um especialista em narrativa.
Reescreva o texto mudando o ponto de vista.
"""

user_prompt = """
Transforme de primeira para terceira pessoa:

"Eu viajei para o Jap√£o pela primeira vez em 2023 e fiquei impressionado
com a mistura de tradi√ß√£o e tecnologia em T√≥quio."
"""

response_perspective = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.4
)

print("\n=== RESULTADO (Perspectiva) ===")
print(response_perspective.choices[0].message.content)
```
