```python
from dotenv import load_dotenv
from groq import Groq
import os
load_dotenv()

client = Groq(
    api_key=os.environ.get("GROQ_API"),  # This is the default and can be omitted
)
```

### Sumariza√ß√£o B√°sica
-  Tend√™ncias e desafios no desenvolvimento de jogos indie

```python
system_prompt = """
Voc√™ √© um editor t√©cnico especializado em jogos independentes.
Crie um resumo objetivo com aproximadamente 1/3 do tamanho do texto original,
preservando os pontos centrais e o tom informativo.
"""

user_prompt = """
A cena de jogos independentes segue em crescimento, com ferramentas como engines gratuitas,
marketplaces digitais e financiamento coletivo tornando poss√≠vel a cria√ß√£o por equipes pequenas.
Desafios frequentes incluem descoberta (getting discovered) em lojas saturadas, financiamento
sustent√°vel ap√≥s o lan√ßamento e equil√≠brio entre inova√ß√£o e usabilidade. Muitos est√∫dios optam
por construir comunidades durante o desenvolvimento (early access, betas) para reduzir riscos.
Tend√™ncias recentes mostram √™nfase em experi√™ncias narrativas curtas, jogos como servi√ßo
em modelo leve, e experimentos com integra√ß√£o de √°udio procedural. O mercado tamb√©m aponta
maior valoriza√ß√£o por jogos com est√©tica √∫nica e mec√¢nicas originais, embora isso nem sempre
se traduza em vendas imediatas. Estrat√©gias de monetiza√ß√£o variam: premium, pay-what-you-want,
DLCs pequenas, e microtransa√ß√µes bem posicionadas em jogos multiplayer. Em suma, o sucesso
indie depende de produto de qualidade, comunidade engajada e escolhas estrat√©gicas de lan√ßamento.
"""

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.2
)

print("=== Sumariza√ß√£o B√°sica ===")
print(response.choices[0].message.content)
```

### Sumariza√ß√£o em Bullets
- Relat√≥rio piloto de mobilidade urbana (dados e a√ß√µes)

```python
system_prompt = """
Voc√™ √© um analista de pol√≠ticas p√∫blicas. Gerei um resumo em bullet points claros e hier√°rquicos:
- Destaque resultados (m√©tricas)
- Liste problemas priorit√°rios
- Sugira 3 a√ß√µes imediatas e 2 a√ß√µes estruturais
Mantenha frases curtas.
"""

user_prompt = """
Relat√≥rio Piloto: Nova rota de ciclovia e corredores de √¥nibus r√°pidos implementada por 12 semanas.
Resultados preliminares: aumento no uso de bicicleta em 22% nas rotas testadas; redu√ß√£o de tempo m√©dio de deslocamento em √¥nibus local em 8%; aumento de reclama√ß√µes de comerciantes sobre estacionamento; registro de 3 incidentes leves envolvendo ciclistas.
Observa√ß√µes: infraestrutura tempor√°ria com sinaliza√ß√£o provis√≥ria; campanhas educativas em 4 bairros; or√ßamento limitado para fiscaliza√ß√£o.
"""

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.25
)

print("\n=== Bullet Points (Mobilidade Urbana) ===")
print(response.choices[0].message.content)
```

### Resumo Executivo

```python
system_prompt = """
Voc√™ √© um consultor executivo. Construa um resumo executivo com:
1) S√≠ntese em 2-3 frases
2) Principais m√©tricas (visitantes, receita direta, parcerias)
3) 3 recomenda√ß√µes acion√°veis para o pr√≥ximo ano (priorizadas)
Seja conciso e orientado para decis√£o.
"""

user_prompt = """
Relat√≥rio: Festival "Olhar Curto" (3 dias).
M√©tricas: 4.200 visitantes (inclui sess√µes pagas e gratuitas), receita direta: R$ 180.000 (ingressos + barra),
patroc√≠nios: R$ 70.000, custos operacionais: R$ 150.000.
Impacto local: aumento de 18% no movimento de bares e caf√©s pr√≥ximos durante o evento;
envolvimento de 6 escolas t√©cnicas; 25 curtas exibidos, 8 produ√ß√µes locais.
Desafios: log√≠stica de exibi√ß√£o ao ar livre (chuva reprogramou 2 sess√µes), bilheteria online com alta taxa de desist√™ncia.
"""

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.15
)

print("\n=== Resumo Executivo (Festival de Curtas) ===")
print(response.choices[0].message.content)
```

### Compara√ß√£o de Estilos de Sumariza√ß√£o
- Resumo t√©cnico (formal) e um resumo curto para redes sociais (casual)

```python
base_text = """
Projeto de conserva√ß√£o marinha no litoral: monitoramento de recifes por sensores ac√∫sticos e foto-identifica√ß√£o de tartarugas;
dados coletados trimestralmente mostram recupera√ß√£o gradual de popula√ß√µes locais (+12% em 2 anos),
mas aumento de res√≠duos pl√°sticos nas √°reas de alimenta√ß√£o; parcerias com pescadores locais reduziram
incidentes de emalhe em 35%. Necessidade de financiamento para ampliar esta√ß√µes de monitoramento
e campanhas comunit√°rias de educa√ß√£o ambiental. Risco clim√°tico: eventos de branqueamento relacionados √† eleva√ß√£o de temperatura da √°gua.
"""

# Resumo t√©cnico
system_prompt_tech = """
Voc√™ √© um redator t√©cnico-cient√≠fico. Produza um resumo formal t√©cnico (3-4 frases),
destacando metodologia, resultados quantitativos e limita√ß√µes.
"""

response_tech = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt_tech},
        {"role": "user", "content": base_text}
    ],
    temperature=0.2
)

# Resumo para redes sociais
system_prompt_social = """
Voc√™ √© um comunicador digital. Fa√ßa uma vers√£o curta e chamativa (1-2 frases) adequada para Instagram,
mantendo precis√£o mas com tom acess√≠vel e motivador.
"""

response_social = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt_social},
        {"role": "user", "content": base_text}
    ],
    temperature=0.45
)
```

```python
print("\n=== Resumo T√©cnico ===")
print(response_tech.choices[0].message.content)
```

```python
print("\n=== Resumo para Redes Sociais ===")
print(response_social.choices[0].message.content)
```

### Sumariza√ß√£o em formato JSON

```python
texto = """
A intelig√™ncia artificial est√° transformando setores como sa√∫de, finan√ßas e educa√ß√£o.
Na medicina, auxilia no diagn√≥stico precoce de doen√ßas. No setor financeiro, detecta fraudes
e otimiza investimentos. Na educa√ß√£o, personaliza o aprendizado para cada estudante.
Apesar dos avan√ßos, existem preocupa√ß√µes √©ticas sobre privacidade, vi√©s algor√≠tmico
e substitui√ß√£o de empregos.
"""

# Prompt pedindo JSON estruturado
json_prompt = f"""
Resuma o seguinte texto em JSON.
O JSON deve conter as chaves:
- "setores" (lista com os setores impactados)
- "beneficios" (lista com os principais benef√≠cios citados)
- "desafios" (lista com os riscos ou preocupa√ß√µes levantadas)

Texto:
{texto}
"""

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[{"role": "user", "content": json_prompt}],
    temperature=0.2
)

# Captura o conte√∫do da resposta
resposta_json = response.choices[0].message.content

print("=== RESUMO EM JSON ===")
print(resposta_json)
```
